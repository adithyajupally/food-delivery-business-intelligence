```python
df_clean=df.copy()
df_clean.head()
```

```python
# Data Cleaning 
df_clean.columns=(
    df_clean.columns.str.lower().str.strip().str.replace(" ","_").str.replace("/","_").str.replace("(","").str.replace(")","")
)
```

```python
df_clean.isnull().sum()
```

```python
# convert data columns 
df_clean['order_placed_at'].value_counts
df_clean["order_placed_at"] = pd.to_datetime(df_clean["order_placed_at"],errors="coerce")
```

```python
#  Create Date Features
df_clean['order_placed_at'].value_counts
df_clean["hour"] = df_clean["order_placed_at"].dt.hour
df_clean["day"] = df_clean["order_placed_at"].dt.day
df_clean["year"] = df_clean["order_placed_at"].dt.year
df_clean["month"] = df_clean["order_placed_at"].dt.month

df_clean["month_name"] = df_clean["order_placed_at"].dt.month_name()
df_clean["day_name"]=df_clean["order_placed_at"].dt.day_name()

df_clean.columns.tolist()
df_clean.shape[1]
```

```python
# Further Data Cleaning
df_clean["instructions"]=df_clean["instructions"].fillna("No Instructions")
df_clean["review"] = df_clean["review"].fillna("No Review")
df_clean["customer_complaint_tag"] = (df_clean["customer_complaint_tag"].fillna("No Complaint"))
df_clean["cancellation___rejection_reason"] = (df_clean["cancellation___rejection_reason"].fillna("No Cancellation"))
df_clean["discount_construct"] = (df_clean["discount_construct"].fillna("No Discount"))
```

```python
df_clean.dtypes
# Only Distance need to be converted to Numeric 
df_clean['distance']=df_clean["distance"].str.replace("km","").str.replace("<1","0.6")
df_clean['distance']=pd.to_numeric(df_clean['distance'])
```

```python
df_clean['distance'].dtype
```

```python
# Handle Missing Operational Values

df_clean["kpt_duration_minutes"] = df_clean["kpt_duration_minutes"].fillna(df_clean["kpt_duration_minutes"].median())

df_clean["rider_wait_time_minutes"] = df_clean["rider_wait_time_minutes"].fillna(df_clean["rider_wait_time_minutes"].median())

```

```python
df_clean.head()
```

```python
# Strip Extra Spaces
text_cols = df_clean.select_dtypes(include='object').columns
for col in text_cols:
    df_clean[col] = df_clean[col].str.strip()
```

```python
# Handling other NaN values
# Don't fill rating with 0. 
df_clean["restaurant_compensation_cancellation"] = (df_clean["restaurant_compensation_cancellation"].fillna(0))
df_clean["restaurant_penalty_rejection"] = (df_clean["restaurant_penalty_rejection"].fillna(0))

```

```python
# Final Code Validation 
print(df_clean.shape)
print(df_clean.isnull().sum())
print(df_clean.duplicated().sum())
```
